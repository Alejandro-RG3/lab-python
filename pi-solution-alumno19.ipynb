```python
import random
import sys
import time

def calc_pi(N):
    M = 0
    for i in range(N):
    # Simulate impact coordinates
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
    # True if impact happens inside the circle
        if x**2 + y**2 < 1.0:
            M += 1
    return 4 * M / N

num_trials = int(sys.argv[1])

initialTime = time.time()
pi = calc_pi(num_trials)
finalTime = time.time()
tiempo_pi=finalTime-initialTime


print("\n \t Computing pi in serial: \n")
print("\t For %d trials, pi = %f\n" % (num_trials,pi))
```

```python
import numpy as np
import time
from numba import njit, prange
import sys

@njit(parallel=True)
def calc_pi_numba(N):
    M=0
    for i in prange(N):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x**2 + y**2 < 1.0:
            M += 1
    return 4 * M / N

pi = calc_pi_numba(num_trials)
```

```python
initialTime = time.time()
pi = calc_pi_numba(num_trials)
finalTime = time.time()
tiempo_pi_numba=finalTime-initialTime
print("Numero Pi con numba:", pi)
print("Tiempo cálculo numba primera ejecución:", tiempo_pi_numba)

#Ejecuto por segunda vez calc_pi_numba
initialTime = time.time()
pi = calc_pi_numba(num_trials)
finalTime = time.time()
tiempo_pi_numba2=finalTime-initialTime
print("Numero Pi con numba (Segunda ejecucion):", pi)
print("Tiempo cálculo numba segunda ejecución:",tiempo_pi_numba2)

print("Mejora de la versión optimizada con respecto a la original:", tiempo_pi/tiempo_pi_numba2,"X")
```
